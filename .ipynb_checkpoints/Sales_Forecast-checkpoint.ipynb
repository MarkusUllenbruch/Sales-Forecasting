# Predict Sales from all Stores of Item 22

### Parameter

```python
ITEM_NR = 22

TIMESTEPS_FORECAST = 10
TIMESTEPS_LAG = 40
```

### Define Training Data to Forecast

```python
def get_sales_per_item(item_idx):
    sales_per_item = train.groupby(['item', 'date'], as_index=False)['sales'].sum()
    sales_per_item = sales_per_item[sales_per_item['item']==item_idx]
    return sales_per_item

train_X = get_sales_per_item(ITEM_NR)
train_X = train_X['sales'].values
train_X
```

```python
len(train_X)
```

### Train/ Test Split

```python
def train_test_split(X_data, train=0.8):
    idx_True = int(X_data.shape[0]*train)
    idx_False = X_data.shape[0] - idx_True
    idx = [True]*idx_True + [False]*idx_False
    idx = np.array(idx)
    
    return X_data[idx], X_data[~idx]

train, test = train_test_split(train_X)

print(train.shape, test.shape)
```

### Standard Scaler

```python
class StandardScaler:
    def __init__(self):
        self.mean = 0.0
        self.std = 0.0
    
    def fit(self, train):
        self.mean = np.mean(train)
        self.std = np.std(train)
    
    def transform(self, data):
        return (data - self.mean) / (self.std + 1e-8)
    
    def retransform(self, data):
        return data * self.std + self.mean
```

```python
scale = True

if scale:
    scaler = StandardScaler()
    scaler.fit(train)
    train = scaler.transform(train)
    test = scaler.transform(test)
```

### Make Supervised Data from Timeseries for LSTM Training

```python
def make_supervised_data(data, t_back, t_forecast):
    train_X, train_y = [], []
    for i in range(t_back, len(data)-t_forecast):
        train_X.append(data[i-t_back:i])
        train_y.append(data[i:i+t_forecast])
    train_X = np.array(train_X)
    train_y = np.array(train_y)
    train_X = train_X.reshape(train_X.shape[0], train_X.shape[1], 1)
    return train_X.astype(np.float32), train_y.astype(np.float32)

X_train, y_train = make_supervised_data(train, TIMESTEPS_LAG, TIMESTEPS_FORECAST)
X_test, y_test = make_supervised_data(test, TIMESTEPS_LAG, TIMESTEPS_FORECAST)
```

```python
print(X_train.shape, y_train.shape)
print(X_test.shape, y_test.shape)
```

[n_samples, seq_len, n_features] for LSTM

Now we are ready to model with LSTM!

## LSTM Modeling

```python
import tensorflow as tf
from tensorflow.keras.layers import LSTM, Dense
```

```python
X_train = tf.convert_to_tensor(X_train, dtype=tf.float32)
y_train = tf.convert_to_tensor(y_train, dtype=tf.float32)
X_test = tf.convert_to_tensor(X_test, dtype=tf.float32)
y_test = tf.convert_to_tensor(y_test, dtype=tf.float32)
```

```python
class LSTM_Model(tf.keras.Model):
    def __init__(self,
                 n_back,      # Timesteps to look back to make prediction
                 n_features,  # Number of features/ Timeseries signals
                 n_forecast): # Timesteps to predict in future
        super().__init__()
        
        self.layer1 = LSTM(50,
                           activation='tanh',
                           input_shape=(n_back, n_features),
                           return_sequences=True)
        
        self.layer2 = LSTM(20,
                           activation='tanh',
                           return_sequences=False)
        
        self.out = Dense(n_forecast)
    
    def call(self, inputs):
        x = self.layer1(inputs)
        x = self.layer2(x)
        x = self.out(x)
        return x
```

```python
model = LSTM_Model(n_back=TIMESTEPS_LAG,
                   n_features=1,
                   n_forecast=TIMESTEPS_FORECAST)

model.compile(optimizer='adam',
              loss=tf.keras.losses.Huber(delta=1.0, reduction="auto", name="huber_loss"))
```

```python
from tensorflow.keras.callbacks import EarlyStopping

early_stopping_monitor = EarlyStopping(
                                    monitor='val_loss',
                                    min_delta=0,
                                    patience=10,
                                    verbose=0,
                                    mode='auto',
                                    baseline=None,
                                    restore_best_weights=True)
```

```python
history = model.fit(X_train,
                    y_train,
                    epochs=500,
                    shuffle=True,
                    validation_data=(X_test, y_test),
                    callbacks = [early_stopping_monitor])
```

```python
model.summary()
```

```python
plt.plot(history.history['loss'], label='Training Loss')
plt.plot(history.history['val_loss'], label='Validation Loss')
plt.xlabel('Epoch')
plt.ylabel('Huber Loss')
plt.legend()
plt.savefig('./Plots/Loss.png')
plt.show()
```

```python
def plot_pred(pred, true):
    data_pred, data_true = [], []
    for i in range(pred.shape[0]):
        a = pred[i][-1]
        b = true[i][-1]
        data_pred.append(a)
        data_true.append(b)
    plt.plot(data_pred)
    plt.plot(data_true)
    plt.show()

    
test_pred = model.predict(X_test)
test_pred = scaler.retransform(test_pred)
test_y = scaler.retransform(y_test)

plot_pred(test_pred, test_y)
```

```python
def plot_predictions(train, test, test_data):
    train_pred = model.predict(test_data)
    
    temp = []
    for i in range(train_pred.shape[0]-1):
        temp.append(train_pred[i, 0])
        
    for value in list(train_pred[-1, :]):
        temp.append(value)
    
    t_pred = np.arange(train.shape[0]+TIMESTEPS_LAG, train.shape[0]+TIMESTEPS_LAG+len(temp))        
    t_train = np.arange(train.shape[0])
    t_test = np.arange(train.shape[0], train.shape[0]+test.shape[0])
    plt.figure(figsize=(14,8))
    plt.plot(t_train, train, label='Train Data')
    plt.plot(t_test, test, '--', label='Test Data')
    plt.plot(t_pred, temp, label='Predictions')
    plt.legend()
    plt.savefig('./Plots/Predictions.png')
    plt.show()
    
plot_predictions(train, test, X_test)
```
